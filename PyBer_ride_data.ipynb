```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
import matplotlib.colors as mcolors
```

```python
# Import csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='barh')
plt.gca().invert_yaxis()
plt.show()
```

```python
errors = pyber_ride_df['Avg. Fare ($USD)'].std() 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", 
                   kind='bar', yerr=errors, ecolor="darkslategray",capsize=3, color="powderblue")
plt.xlabel('Month', fontsize=13)
plt.title("Average Fare Cost per Month with Std. Deviation", fontsize=13)
labelpad=15
plt.show()
```
